# Group by in pandas 
**Definition :** Group by is used to divide data in groups  on the basis of categorical columns

```python
# synatx : dataframe.groupby('categorical_column')
gb=df.groupby('Hunter_status')
gb['Animal_weight'].sum()
```

## Functions of a group by 

```python
# len() : len function is used to find the total number of groups in group by
len(gb)
```

```python
# size() : size is used to find the number of items in each group
gb.size()
```

```python
#first() / last() : these functions are used to find the first and last values of a group by object
gb.first()
gb.last()
```

```python
# nth() : it is used to find the nth value of group by 
gb.nth(2)
```

```python
# get_groups() : this function is used to get all the values related to a specfic group of group by 
gb.get_group('Poor')
```

```python
#group() : this funcction is used to get all the groups of a group by object
gb.groups
```

```python
# agg() : agg stands for aggregate ,is a special type of fuction of group by which performs more than one aggregate function at a time
# syntax: groupbyobj.agg({ 'colm1':'aggfunction','colm2':'aggfunction'})
gb.agg({'Animal_weight':'mean','Hunter_status':'count'})
```
