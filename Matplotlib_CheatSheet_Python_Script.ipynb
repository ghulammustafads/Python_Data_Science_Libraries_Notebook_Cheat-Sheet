```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Optional Styling
plt.style.use('seaborn-v0_8')
```

```python
# Reproducibility
np.random.seed(42)

# Create sample dataset
data = {
    "Date": pd.date_range(start="2023-01-01", periods=100),
    "Sales": np.random.randint(100, 500, 100),
    "Marketing_Spend": np.random.randint(50, 300, 100),
    "Profit": np.random.randint(20, 200, 100),
    "Region": np.random.choice(["North", "South", "East", "West"], 100),
    "Customer_Age": np.random.randint(18, 65, 100)
}

df = pd.DataFrame(data)
df.head()
```

## plt.plot()

### Definition
`plt.plot()` is used to create line plots. It is commonly used to visualize trends over continuous data such as time series.

### Syntax
plt.plot(x, y, color=None, linestyle=None, marker=None, linewidth=None, label=None)

### Important Parameters
- **x**: Values for x-axis
- **y**: Values for y-axis
- **color**: Line color
- **linestyle**: '-', '--', '-.', ':'
- **marker**: 'o', 's', '^', '*'
- **linewidth**: Thickness of line
- **label**: Legend label

### When to Use
- Time series analysis
- Trend visualization
- Comparing multiple continuous variables

```python
plt.figure(figsize=(8,5))
plt.plot(df["Date"], df["Sales"],color="blue", linestyle="-", marker="o",linewidth=2,label="Sales")

plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## plt.scatter()

### Definition
`plt.scatter()` creates a scatter plot used to visualize relationships between two numerical variables.

### Syntax
plt.scatter(x, y, s=None, c=None, alpha=None, marker=None)

### Important Parameters
- **x, y**: Numeric variables
- **s**: Marker size
- **c**: Color
- **alpha**: Transparency (0 to 1)
- **marker**: Shape of marker

### When to Use
- Correlation analysis
- Detecting outliers
- Pattern identification

## plt.bar()

### Definition
`plt.bar()` creates vertical bar charts for categorical comparisons.

### Syntax
plt.bar(x, height, color=None, width=0.8)

### Important Parameters
- **x**: Categories
- **height**: Values
- **color**: Bar color
- **width**: Width of bars

### When to Use
- Comparing categories
- Aggregated statistics visualization

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(6,5))
plt.bar(region_sales.index, region_sales.values,color="orange", width=0.6)

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

## plt.hist()

### Definition
`plt.hist()` is used to display the distribution of numerical data.

### Syntax
plt.hist(data, bins=None, color=None, edgecolor=None)

### Important Parameters
- **data**: Numeric data
- **bins**: Number of intervals
- **color**: Fill color
- **edgecolor**: Border color

### When to Use
- Understanding distribution
- Checking skewness
- Data exploration

```python
plt.figure(figsize=(6,5))
plt.hist(df["Customer_Age"], bins=10, color="skyblue",  edgecolor="black")

plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## plt.boxplot()

### Definition
`plt.boxplot()` visualizes the distribution of data through quartiles and highlights outliers.

### Syntax
plt.boxplot(data, vert=True, patch_artist=False)

### Important Components
- Median (center line)
- Q1 & Q3 (box edges)
- Whiskers
- Outliers

### When to Use
- Detecting outliers
- Comparing distributions

```python
plt.figure(figsize=(6,5))
plt.boxplot(df["Sales"], vert=True)

plt.title("Sales Distribution")
plt.ylabel("Sales")
plt.show()
```

## plt.pie()

### Definition
`plt.pie()` creates a pie chart to show proportional distribution.

### Syntax
plt.pie(x, labels=None, autopct=None, startangle=0)

### Important Parameters
- **x**: Values
- **labels**: Category labels
- **autopct**: Percentage format
- **startangle**: Rotation angle

### When to Use
- Showing proportions
- Percentage contribution

```python
region_counts = df["Region"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(region_counts.values, labels=region_counts.index, autopct="%1.1f%%", startangle=90)

plt.title("Region Distribution")
plt.show()
```

## plt.subplots()

### Definition
`plt.subplots()` creates multiple plots within a single figure.

### Syntax
fig, ax = plt.subplots(nrows, ncols, figsize=(width, height))

### Important Parameters
- **nrows**: Number of rows
- **ncols**: Number of columns
- **figsize**: Figure size

### When to Use
- Dashboard-style layouts
- Comparing multiple visualizations

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))

ax[0,0].plot(df["Sales"])
ax[0,0].set_title("Sales")

ax[0,1].scatter(df["Marketing_Spend"], df["Sales"])
ax[0,1].set_title("Marketing vs Sales")

ax[1,0].hist(df["Customer_Age"])
ax[1,0].set_title("Age Distribution")

ax[1,1].boxplot(df["Profit"])
ax[1,1].set_title("Profit Distribution")

plt.tight_layout()
plt.show()
```

## plt.savefig()

### Definition
`plt.savefig()` saves the current figure to a file.

### Syntax
plt.savefig(filename, dpi=None, bbox_inches=None)

### Important Parameters
- **filename**: File name with extension
- **dpi**: Resolution
- **bbox_inches='tight'**: Removes extra whitespace

### When to Use
- Saving plots for reports
- Publishing images

```python
plt.figure(figsize=(6,5))
plt.plot(df["Profit"])

plt.title("Profit Plot")
plt.savefig("profit_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

## plt.annotate()

### Definition
`plt.annotate()` adds text annotations to a plot.

### Syntax
plt.annotate(text, xy, xytext, arrowprops=None)

### Important Parameters
- **text**: Annotation text
- **xy**: Point to annotate
- **xytext**: Text position
- **arrowprops**: Arrow styling

### When to Use
- Highlighting peaks
- Marking important points

```python
plt.figure(figsize=(6,5))
plt.plot(df["Sales"])

plt.annotate("Maximum Sales",xy=(df["Sales"].idxmax(), df["Sales"].max()), xytext=(20, 450),  arrowprops=dict(facecolor="black"))

plt.title("Sales with Annotation")
plt.show()
```

## plt.yscale()

### Definition
`plt.yscale()` changes the scale of y-axis.

### Syntax
plt.yscale("log")

### Available Options
- "linear"
- "log"
- "symlog"

### When to Use
- Exponential growth data
- Large value differences

```python
plt.figure(figsize=(6,5))
plt.plot(df["Sales"])

plt.yscale("log")
plt.title("Log Scale Sales Plot")
plt.show()
```
